```python
from data_download import download_data
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

# Wine Dataset

```python
X_wine, y_wine = download_data("Wine")
```

```python
X_wine
```

```python
y_wine = y_wine['class'] - 1
```

```python
sns.countplot(x=y_wine)

plt.xlabel('Class')
plt.ylabel('Frequency')
plt.title('Class Distribution in Wine Dataset')
plt.show()
```

```python
X_wine.to_csv('data/X_wine', index=False)
y_wine.to_csv('data/y_wine', index=False)
```
